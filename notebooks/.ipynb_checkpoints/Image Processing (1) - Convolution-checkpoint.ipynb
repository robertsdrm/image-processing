## Convolution ##


Convolution is a one of the core tools in image processing, enabling basic functions such as smoothing, sharpening, and edge extraction. It works by running a kernel over the image. The kernel is typically a 3x3 or 5x5 matrix that is centred on the pixel it's operating on. The new value for the pixel is determined as a function of the kernel and the pixels surrounding the image pixel that's being operated on. 

The intention is to avoid using libraries to ensure that the main functions are as clear as possible. PIL is used solely to load images and matplotlib to display the loaded and the processed output. Numpy is used for array manipulation.

Define kernels to use in the convolution, in those included below the linear and Gaussian kernels will smooth (blur) the image. A Sobel filter is used to extract edges.

```python
kernel_lin5 = [1/25]*25
kernel_laplacian5 = [-4,-4,0,-1,-4,
                     -1,2,3,3,-1,
                     0,3,4,3,0,
                     -1,2,3,3,-1,
                     -4,-4,0,-1,-4]
kernel_sobel5 = [1,2,0,-2,-1,4,8,0,-8,4,6,12,0,-12,-6,4,8,0,-8,4,1,2,0,-2,-1]
kernel_sharp5 = [-0.00391,-0.01563,-0.02344,-0.01563,-0.00391,
                 -0.01563,-0.06250,-0.09375,-0.06250,-0.01563,
                 -0.02344,-0.09375,1.85980,-0.09375,-0.02344,
                 -0.01563,-0.06250,-0.09375,-0.06250,-0.01563,
                 -0.00391,-0.01563,-0.02344,-0.01563,-0.00391]
#TODO - write function to generate arbitrary Gaussian kernel
kernel_gauss5 = [0.003765,0.015019,0.023792,0.015019,0.003765,
                 0.015019,0.059912,0.094907,0.059912,0.015019,
                 0.023792,0.094907,0.153042,0.094907,0.023792,
                 0.015019,0.059912,0.094907,0.059912,0.015019,
                 0.003765,0.015019,0.023792,0.015019,0.003765]
```

Given the image, a kernel and a pixel in the image, returns the new value of the pixel after applying the kernel

```python
def get_window(img_array, p, kernel):
    #TODO: Improve this by pre-computing the offsets
    offsets = [(p-(d2*2)-2),(p-(d2*2)-1),(p-(d2*2)),(p-(d2*2)+1),(p-(d2*2)+2),
               (p-d2-2),(p-d2-1),(p-d2),(p-d2+1),(p-d2+2),
               (p-2),(p-1),p,(p+1),(p+2),
               (p+d2-2),(p+d2-1),(p+d2),(p+d2+1),(p+d2+2),
               (p+(d2*2)-2),(p+(d2*2)-1),(p+(d2*2)),(p+(d2*2)+1),(p+(d2*2)+2)]
    
    new_pixel = np.zeros(3,dtype=np.uint8)
    for i in range(0,len(offsets)):
        if offsets[i] >= 0 and offsets[i] < len(img_array):
            new_pixel[0] = new_pixel[0] + (img_array[offsets[i]][0]*kernel[i])
            new_pixel[1] = new_pixel[1] + (img_array[offsets[i]][1]*kernel[i])
            new_pixel[2] = new_pixel[2] + (img_array[offsets[i]][2]*kernel[i])
    return new_pixel
```

Select a kernel and apply it to each pixel in the image. There are likely to be far more efficient ways to do this. This implementation is very slow and degrades significantly as the image size increases. Setting the number of passes will apply the same convolution repeatedly.

```python
num_passes = 1
kernel = kernel_laplacian5
if kernel == kernel_sobel5 or kernel == kernel_laplacian5:
    array_1d = convert2grey(array_1d)
no_pixels = len(array_1d)
convolved = [np.zeros(3,dtype=np.uint8)]*len(array_1d)
for i in range(0, num_passes):
    print("Pass "+ str(i+1)+ ": Pixels to process: "+str(no_pixels))
    convolved = [get_window(array_1d, pixel, kernel) for pixel,v in enumerate(array_1d)]
    print("Pass " + str(i+1) + ": Pixels processed: "+str(len(convolved)))
```

Convert the flattened image back to the correct shape, i.e. heigh, width, and depth (number of channels). Display the resulting image inline.

```python
convolved_img = np.asarray(convolved).reshape(d1,d2,d3)
%matplotlib inline
img_array = np.asarray(img)
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111)
ax.imshow(convolved_img)
print("Convolved image shape (y,x,channels): " + str(convolved_img.shape))
```
